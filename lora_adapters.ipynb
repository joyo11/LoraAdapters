```python
METRICS_DIR_NAME = "./evaluations/"
```

```python
!mkdir {METRICS_DIR_NAME}
```

```python
login(token=os.environ["HF_TOKEN"])

def upload_folder_to_HF (folder_name):
  upload_folder(folder_path=folder_name,
      repo_id="ahmedashrafay/staradapters-evaluations",
      commit_message=f"Add metrics/generations for {MODEL_NAME}")
```

```python
def run_evaluation (MODEL_NAME, TASK):
  FILE = f"{MODEL_NAME.split('/')[-1].replace('-', '_')}_{TASK.replace('-', '_')}_metrics.json"
  METRICS_FILE_NAME = METRICS_DIR_NAME + FILE

  !accelerate launch main.py \
          --model {MODEL_NAME} \
          --task {TASK} \
          --n_samples 15 \
          --batch_size 10\
          --temperature 0.2 \
          --metric_output_path {METRICS_FILE_NAME} \
          --allow_code_execution
```

# Evaluations

```python
run_evaluation (MODEL_NAME, "humaneval")
upload_folder_to_HF (METRICS_DIR_NAME)
```
